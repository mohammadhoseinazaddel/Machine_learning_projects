فاصله ها را استفاده کرده و از منیفولد استفاده میکند

روی دیتای ماه آن را تست کنید و تغییرات را با پلات کردن نشان دهید

```python
from sklearn.decomposition import FastICA
from sklearn.manifold import Isomap
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.01)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
ica = FastICA(n_components=1, random_state=123)
ica.fit(X)
ica_result = ica.transform(X)
ica_result
```

```python
plt.scatter(ica_result[:, 0], range(0, len(ica_result)), c=y)
```

```python
isomap = Isomap(n_components=1)
isomap.fit(X)
isomap_result = isomap.transform(X)
isomap_result
```

```python
plt.scatter(isomap_result[:, 0], range(0, len(isomap_result)), c=y)
```

```python
import umap
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, noise=0.01)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
tsne = TSNE(n_components=1, perplexity=50 ,random_state=123)
tsne_result = tsne.fit_transform(X)
tsne_result
```

```python
plt.scatter(tsne_result[:, 0], range(0, len(tsne_result)), c=y)
```

```python
u = umap.UMAP(n_components=1)
u.fit(X)
u_result = u.transform(X)
u_result
```

```python
plt.scatter(u_result[:, 0], range(0, len(u_result)), c=y)
```
