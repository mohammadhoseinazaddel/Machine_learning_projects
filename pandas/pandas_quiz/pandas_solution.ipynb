# solution

```python
import pandas as pd
```

```python
data = pd.read_csv('quize_pandas.zip', compression='zip')
```

```python
pm = (data
 .groupby('Product Model'.split())
 .sum()
)
```

```python
pm
```

```python
cpm = (data
 .groupby('Customer Product Model'.split())
 .sum()
)
```

```python
cpm_p = (cpm / pm * 100).round(-1)
```

```python
cpm_p
```

```python
cpm_p_table = (cpm_p
               .unstack([0, 1])
               .dropna(axis=1, how='all')
              )
```

```python
cpm_p_table
```

```python
ordered_list = pd.merge(
    left=data, right=pm, how='left', 
    left_on='Product Model'.split(), right_index=True, 
    suffixes=(None, '_total')
)
```

```python
ordered_list_sorted = (
    ordered_list
    .set_index('Customer Product Model'.split())
    .sort_index()
)
```
